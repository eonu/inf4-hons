```python
import os, glob
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
labels = ('nd', 'mnd', 'fu', 'fd', 'sh', 't', 'ti')

def plot(model, number):
    file = os.path.join('experiments', model, 'final test {}.npy'.format(number))
    df = pd.DataFrame(np.load(file), index=labels, columns=labels)
    plt.figure(figsize=(6, 4.5))
    sns.heatmap(df, annot=True, cmap='PuBu')
    plt.yticks(rotation=0) 
    plt.xlabel('Predicted class')
    plt.ylabel('Actual class')
    # Fix for matplotlib bug that cuts off top/bottom of seaborn visualizations
    b, t = plt.ylim()
    plt.ylim(b + 0.5, t - 0.5)
    plt.tight_layout()
#     plt.savefig('confusion.pdf')
    plt.show()
```

```python
for clf in ('knn', 'hmm', 'ffnn', 'lstm'):
    plot(clf, 4)
```
