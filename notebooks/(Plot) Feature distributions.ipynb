```python
import os, sys
import numpy as np, matplotlib.pyplot as plt, pandas as pd, seaborn as sns, sequentia as seq

# Import utility functions and classes
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from utils import *

# ggplot style
plt.style.use('ggplot')
```

```python
fields = ['Rx', 'Ry', 'Rz']
```

```python
# Load the MoCap dataset
loader = MoCapLoader(normalized=False)
X, _ = loader.load(fields)
```

```python
Xc = np.concatenate(X)
fig, axs = plt.subplots(1, 3, figsize=(15, 3.5))
sns.distplot(Xc[:, 0], bins=75, color='r', ax=axs[0], kde=False).set_xlabel('$\\varphi_x^{(t)}$', rotation=0, fontsize=14)
axs[0].set_ylabel('Frequency')
sns.distplot(Xc[:, 1], bins=75, color='g', ax=axs[1], kde=False).set_xlabel('$\\varphi_y^{(t)}$', rotation=0, fontsize=14)
sns.distplot(Xc[:, 2], bins=75, color='b', ax=axs[2], kde=False).set_xlabel('$\\varphi_z^{(t)}$', rotation=0, fontsize=14)
plt.tight_layout()
plt.savefig('plots/feature-distributions.pdf')
plt.show()
```
