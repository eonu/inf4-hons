```python
import os, sys
import numpy as np, matplotlib.pyplot as plt, pandas as pd, seaborn as sns

# Import utility functions and classes
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from utils import *

# ggplot style
plt.style.use('ggplot')
```

```python
fields = ['Rx', 'Ry', 'Rz', 'Tx', 'Ty', 'Tz']
```

```python
# Load the MoCap dataset
loader = MoCapLoader(normalized=False)
X, _ = loader.load(fields)
```

```python
plt.figure(figsize=(7, 4))
plt.hist([len(x) for x in X], bins=75)
plt.xlabel('Duration (frames)')
plt.ylabel('Count (log)')
plt.yscale('log')
plt.tight_layout()
plt.savefig('Plots/durations.pdf')
plt.show()
```

```python
plt.figure(figsize=(7, 4))
plt.hist([len(x) for x in BinDownsample(bin_size=50).transform(X)], bins=75)
plt.xlabel('Duration (frames)')
plt.ylabel('Count (log)')
plt.yscale('log')
plt.tight_layout()
plt.savefig('Plots/bin-durations.pdf')
plt.show()
```
