```python
import os, glob
import numpy as np, pandas as pd, matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
def get_results(clf, type_):
    results = {'accuracy': {}}
    
    for file in glob.glob(os.path.join('experiments', 'preprocess', 'downsample', clf, '*')):
        if type_ in file:
            window_size = int(os.path.basename(file).split('{} validation '.format(type_))[0])
            with open(file, 'r') as f:
                for line in f:
                    key, value = line.strip('\n').split(': ')
                    value = float(value)
                    if key in results.keys():
                        if window_size not in results[key]:
                            results[key][window_size] = [value]
                        else:
                            results[key][window_size].append(value)                        
                    
    # Convert to numpy array
    results = {k1:{k2:np.array(d2) for k2, d2 in d1.items()} for k1, d1 in results.items()}
        
    # Calculate means and stds
    means = {k1:{k2:np.mean(d2) for k2, d2 in d1.items()} for k1, d1 in results.items()}
    stds = {k1:{k2:np.std(d2)/np.sqrt(5) for k2, d2 in d1.items()} for k1, d1 in results.items()}
                    
    return means, stds
```

```python
display(get_results('hmm', 'decimate'))
display(get_results('ffnn', 'decimate'))
```

```python
display(get_results('hmm', 'bin'))
display(get_results('ffnn', 'bin'))
```
