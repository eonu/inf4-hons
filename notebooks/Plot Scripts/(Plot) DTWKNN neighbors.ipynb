```python
import os, glob
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
N = 5 # Number of experiments
K_RANGE = (1, 5, 10, 15, 20, 25, 30) # Range of K
K_SETTINGS = len(K_RANGE) # Number of K settings

results = {i:{
    'accuracy': np.zeros(N), 
    'f1': np.zeros(N), 
    'memory_fit': np.zeros(N),
    'time_fit': np.zeros(N),
    'memory_predict': np.zeros(N),
    'time_predict': np.zeros(N)
} for i in K_RANGE}

fields = list(results[1].keys())

for file in glob.glob(os.path.join('experiments', 'knn', '*')):
    if ('validation' not in file) or (file.endswith('.npy')) or ('neighbors' not in file):
        continue
    else:
        name = os.path.basename(file)
        n_states, repeat = [int(part) for part in name.split('neighbors validation ')]
        with open(file, 'r') as f:
            for line in f:
                key, value = line.strip('\n').split(': ')
                if key in fields:
                    results[n_states][key][repeat] = float(value)
```

```python
def plot(results, field, xlabel, ylabel, ylim=None, color=None, figsize=None):
    means = np.zeros(K_SETTINGS)
    errors = np.zeros(K_SETTINGS)
    for i, k in enumerate(K_RANGE):
        data = results[k][field]
        means[i] = np.mean(data)
        errors[i] = np.std(data) / np.sqrt(N)
    
    labels = [str(i) for i in K_RANGE]
    x_pos = np.arange(K_SETTINGS)
    
    # Build the plot
    fig, ax = plt.subplots(figsize=figsize)
    ax.bar(x_pos, means, yerr=errors, align='center', alpha=0.7, color=color, ecolor='black', capsize=10, error_kw=dict(lw=1, capsize=3, capthick=1))
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_xticks(x_pos)
    ax.set_xticklabels(labels)
    ax.set_ylim(ylim)

    # Save the figure and show
    plt.tight_layout()
    plt.savefig('{}.pdf'.format(field))
    plt.show()
```

```python
plot(results, 'accuracy', 'Number of neighbors ($k$)', 'Accuracy', ylim=(0.38, 0.48))
```

```python
plot(results, 'time_fit', 'Number of neighbors ($k$)', 'Time to fit (s)', color='#5ea4bf', figsize=(5, 3), ylim=(0.06, 0.095))
```

```python
plot(results, 'time_predict', 'Number of neighbors ($k$)', 'Time to predict (s)', color='#5ea4bf', figsize=(5, 3))
```

```python
plot(results, 'memory_fit', 'Number of neighbors ($k$)', 'Peak memory\nduring fitting (MiB)', ylim=(295, 301), color='#417353', figsize=(5, 3))
```
