```python
import os, sys
import numpy as np, matplotlib.pyplot as plt, pandas as pd, seaborn as sns
from sequentia import *

# Import utility functions and classes
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from utils import *

# ggplot style
plt.style.use('ggplot')
```

```python
fields = ['Rx', 'Ry', 'Rz', 'Tx', 'Ty', 'Tz']
```

```python
# Load the MoCap dataset
loader = MoCapLoader(normalized=False)
X, y = loader.load(fields)
```

```python
x = X[20][:, 2]
x10 = Filter(window_size=10, method='median').transform(x)
x25 = Filter(window_size=25, method='median').transform(x)

fig, axs = plt.subplots(1, 3, figsize=(15, 3), sharey=True)
axs = axs.flat
axs[0].plot(x)
axs[0].set_title('Original signal', fontsize=14)
axs[0].set_ylabel('$\\varphi^{(t)}_z$', rotation=0, fontsize=16, labelpad=12)
axs[0].set_xlabel('Duration (frames)', fontsize=12)
axs[1].plot(x10)
axs[1].set_title('Filtered signal ($n=10$)', fontsize=14)
axs[1].set_xlabel('Duration (frames)', fontsize=12)
axs[2].plot(x25)
axs[2].set_title('Filtered signal ($n=25$)', fontsize=14)
axs[2].set_xlabel('Duration (frames)', fontsize=12)
plt.tight_layout()
plt.savefig('Plots/filtering.pdf')
plt.show()
```
